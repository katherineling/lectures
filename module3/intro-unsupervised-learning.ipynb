```python
%matplotlib inline
import pandas as pd
import numpy as np
census = pd.read_csv('./acs2015_county_data.csv')
census.describe()
```

```python
from sklearn.cluster import KMeans
kmeans1 = KMeans(n_clusters=3)
kmeans2 = KMeans(n_clusters=4)
kmeans3 = KMeans(n_clusters=6)
```

```python
census_clusters1 = kmeans1.fit(censusC)
census_clusters2 = kmeans2.fit(censusC)
census_clusters3 = kmeans3.fit(censusC)
kmeans = KMeans(n_clusters=4)
census_clusters = kmeans.fit(censusC)
# census_clusters.cluster_centers_
```

```python
from sklearn.decomposition import PCA
census_numeric = census._get_numeric_data()
census_numeric.dropna(how='any', inplace = True)
census_numeric = census_numeric.reset_index()
pca = PCA(n_components=2).fit(census_numeric)
pca_2d = pca.transform(census_numeric)
```

```python
import matplotlib.pyplot as pl
for i in range(0, pca_2d.shape[0]):
    if census_numeric['ClusterNumber'][i] == 0.0:
        c1 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='r',marker='+')
    elif census_numeric['ClusterNumber'][i] == 1.0:
        c2 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='g',marker='o')
    elif census_numeric['ClusterNumber'][i] == 2.0:
        c3 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='b',marker='*')
    elif census_numeric['ClusterNumber'][i] == 3.0:
        c4 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='y',marker='+')   
pl.legend([c1, c2, c3, c4], ['Cat1', 'Cat2','Cat3','Cat4'])
pl.title('Out')
pl.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

census_sample = censusC.sample(n=10)
hier_clust = AgglomerativeClustering(linkage='ward')
census_hier = hier_clust.fit(census_sample)
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
    
plot_dendrogram(census_hier, labels=census_hier.labels_)
```
