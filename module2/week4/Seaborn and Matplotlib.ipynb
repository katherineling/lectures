### Basic Plots

Line Plots

```python
x1 = np.arange(0,10,1)
print(x1)
len(x1)
```

```python
x2 = np.linspace(0,10,10)
print(x2)
```

```python
plt.plot(x1, x2, label='linear')
plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Linear vs Non Linear Plots")
plt.legend()

plt.show()
```

QUICK QUIZ 

PLot a cosine curve with labels and title

```python
x = np.arange(1,10,.1)
y = np.cos(x)
plt.plot(x,y, label = "cosine curve")
plt.xlabel('x')
plt.ylabel('cosine')
plt.legend()
plt.show()
```

Making subplots 

```python
fig, [ax1, ax2] = plt.subplots(2,1)
ax1.plot(x1,x2)
ax1.set_title('LINEAR PLOT')   #Note how we are using set_ for subplots
ax1.set_xlabel('x label')      #Note how we are using set_ for subplots
ax1.set_ylabel('y label')      #Note how we are using set_ for subplots 

ax2.plot(x,y)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```python
def my_plots(ax, x,y, params):
    return(ax.plot(x,y,**params))

fig, (ax1, ax2) = plt.subplots(1, 2)
my_plots(ax1, x1, x2, {'marker': '*'})
my_plots(ax2, x, y, {'marker': 'o'})

ax1.set_title('LINEAR PLOT')
ax2.set_title('COSINE PLOT')
```

## Muliple axis same plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color='red')
ax1.plot(t, data1, color='red')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('sin', color='b') 
ax2.plot(t, data2, color='b')
ax2.tick_params(axis='y')

plt.show()
```

Pie Charts

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]
explode = (0, 0.05, 0)  # only "explode" the 2nd slice (i.e. 'Non-Vegetarians')
fig1, [ax1,ax2] = plt.subplots(1,2)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax2.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
# ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Seaborn

```python
import seaborn as sns
import pandas as pd
# sns.set()
```

Line Plot

```python
data1 = pd.DataFrame()
data1['X'] = np.arange(0,10,1)
data1['Y'] = data1['X']**2
data1.head()
```

```python
sns.lineplot(data1.X, data1.Y)
```

Scatter Plot using lmplot() and scatterplot()

1. lmplot()

```python
sns.lmplot("X", "Y", data=data1, fit_reg=True)
```

As you can see, here it is also possible to enhance a scatterplot to include a linear regression model
