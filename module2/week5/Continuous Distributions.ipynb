```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1, 100)

ax.plot(x, uniform.pdf(x))
```

```python
fig, ax = plt.subplots(1, 1)
a=1
b=3
x = np.linspace(a,b, 100)
y = uniform.pdf(x, a, b)
ax.plot(x, y)
```

```python
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

```python
import math
from scipy.stats import norm
normal = norm(0, math.sqrt(9)) 
print(normal.pdf(4))
print(normal.cdf(2))       
print(normal.rvs()) 
fig, ax = plt.subplots(1, 1)
x = np.linspace(-3,3,1000)
y = norm.pdf(x)
ax.plot(x,y)
plt.show()
ax.hist(norm.rvs(100), density=True, histtype='stepfilled', alpha=0.2)
plt.show()
```

```python
from scipy.stats import expon
exp = expon(1) 
print(exp.pdf(4))
print(exp.cdf(2))       
print(exp.rvs()) 
x = np.linspace(0,10,100)
y = expon.pdf(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x,y)
```
